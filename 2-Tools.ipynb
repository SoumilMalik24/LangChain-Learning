## Tools

```python
from langchain.tools import tool

@tool
def get_weather(loaction:str) -> str:
    """Get the weather for a specific location"""
    return f"The weather for {loaction}."

model_with_tools = model.bind_tools([get_weather])
```

```python
response = model_with_tools.invoke("What is the weather like in bangalore?")
print(response)
```

```python
response.tool_calls
```

### Tool execution loop

```python
# Step1: Model generate tool call
messages = [
    {"role": "user", "content": "What is the weather like in bangalore?"}
]
ai_msg = model_with_tools.invoke(messages)

messages.append(ai_msg)

```

```python
#Step 2 : Execute tools and collect results
for tool_call in ai_msg.tool_calls:

    # Execute the tool with the generated arguements
    tool_result = get_weather.invoke(tool_call)
    messages.append(tool_result)
```

```python
# Step 3 : Pass the results back to the model for the final response
final_response = model_with_tools.invoke(messages)
print(final_response.text)
```
